```python
import pandas as pd
from ipyvizzu import Chart, Data, Config, Style
from ipyvizzustory import Story, Slide, Step

data_frame = pd.read_csv(
    "https://raw.githubusercontent.com/vizzuhq/vizzu-workshops/main/2022-11-26-Pyjamas/data/worldpopulation.csv",
    dtype={"Year": str},
)
data = Data()
data.add_data_frame(data_frame)

display(data_frame)
```

```python
story = Story(data=data)
story.set_feature("tooltip", True)
story.set_size("100%", "100%")

slide1 = Slide(
    Step(
        Config.area(
            {  # stackedColumn (stackedBy), groupedColumn (groupedBy), percentageArea (stackedBy), stream (stackedBy), violin (splittedBy)
                "x": "Year",
                "y": "Medium",
                "title": "World Population 1953-2098",
            }
        ),
        Style({"plot": {"xAxis": {"label": {"angle": 2.0}}}}),
    )
)

story.add_slide(slide1)

slide2 = Slide(
    Step(
        Config.stackedArea(
            {  # stackedColumn (stackedBy), groupedColumn (groupedBy), percentageArea (stackedBy), stream (stackedBy), violin (splittedBy)
                "x": "Year",
                "y": "Medium",
                "stackedBy": "Continent",
                "title": "Population by Continents 1953-2098",
            }
        ),
    )
)
story.add_slide(slide2)

slide3 = Slide()

slide3.add_step(
    Step(
        Data.filter("(record.Continent == 'Europe' )"),
        Config.stackedArea(
            {  # stackedColumn (stackedBy), groupedColumn (groupedBy), percentageArea (stackedBy), stream (stackedBy), violin (splittedBy)
                "x": "Year",
                "y": "Medium",
                "stackedBy": "Continent",
                "title": "Population of Europe 1953-2098",
            }
        ),
    )
)

story.add_slide(slide3)

slide4 = Slide()

slide4.add_step(
    Step(
        Data.filter("(record.Continent == 'Europe' )"),
        Config.stackedArea(
            {  # stackedColumn (stackedBy), groupedColumn (groupedBy), percentageArea (stackedBy), stream (stackedBy), violin (splittedBy)
                "x": "Year",
                "y": "Medium",
                "stackedBy": "Subregion",
                "title": "Population of Europe 1953-2098",
            }
        ),
    )
)

story.add_slide(slide4)

slide5 = Slide()

slide5.add_step(
    Step(
        Data.filter("(record.Subregion == 'Western Europe')"),
        Config.stackedArea(
            {  # stackedColumn (stackedBy), groupedColumn (groupedBy), percentageArea (stackedBy), stream (stackedBy), violin (splittedBy)
                "x": "Year",
                "y": "Medium",
                "stackedBy": "Subregion",
                "title": "Population of Western Europe 1953-2098",
            }
        ),
    )
)

story.add_slide(slide5)

slide6 = Slide()

slide6.add_step(
    Step(
        Data.filter("(record.Subregion == 'Western Europe')"),
        Config.stackedArea(
            {  # stackedColumn (stackedBy), groupedColumn (groupedBy), percentageArea (stackedBy), stream (stackedBy), violin (splittedBy)
                "x": "Year",
                "y": "Medium",
                "stackedBy": "Country",
                "title": "Population of Western Europe by Country 1953-2098",
            }
        ),
    )
)

story.add_slide(slide6)

slide7 = Slide()

slide7.add_step(
    Step(
        Data.filter("(record.Country == 'Germany')"),
        Config.stackedArea(
            {  # stackedColumn (stackedBy), groupedColumn (groupedBy), percentageArea (stackedBy), stream (stackedBy), violin (splittedBy)
                "x": "Year",
                "y": "Medium",
                "stackedBy": "Country",
                "title": "Population of Germany with Medium prediction 1953-2098",
            }
        ),
    )
)

story.add_slide(slide7)

slide8 = Slide()

slide8.add_step(
    Step(
        Data.filter("(record.Country == 'Germany')"),
        Config.stackedArea(
            {  # stackedColumn (stackedBy), groupedColumn (groupedBy), percentageArea (stackedBy), stream (stackedBy), violin (splittedBy)
                "x": "Year",
                "y": "Low",
                "stackedBy": "Country",
                "title": "Population of Germany with Low prediction 1953-2098",
            }
        ),
    )
)

story.add_slide(slide8)

slide9 = Slide()

slide9.add_step(
    Step(
        Data.filter("(record.Country == 'Germany')"),
        Config.stackedArea(
            {  # stackedColumn (stackedBy), groupedColumn (groupedBy), percentageArea (stackedBy), stream (stackedBy), violin (splittedBy)
                "x": "Year",
                "y": "High",
                "stackedBy": "Country",
                "title": "Population of Germany with High prediction 1953-2098",
            }
        ),
    )
)

story.add_slide(slide9)


story.play()
```
